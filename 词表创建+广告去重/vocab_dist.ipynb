```python
import re
import numpy as np
from matplotlib import pyplot as plt
from collections import defaultdict
%matplotlib inline
```

```python
#加载词表，只使用词频特征
pattern = re.compile(r'^[\u4e00-\u9fff]+$')
vocab_merge_count = defaultdict(int)
with open('merge_vocab.txt','r',encoding='utf-8') as f:
    for word in f:
        if word[0]!='\t':
            k,v = word.split('\t')
            v = int(v[:-1])
        if len(k)==1 or pattern.match(k):
            vocab_merge_count[k] += v
print('候选词表大小:',len(vocab_merge_count))
```
